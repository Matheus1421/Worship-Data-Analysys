```python
df_exec = pd.read_csv('/content/drive/MyDrive/PROJETOS/DASHBOARD_ADOLESCENTES_REPERTORIO/ADOLESCENTES - EXECUCAO.csv')

```

```python
df_musicas = pd.read_csv('/content/drive/MyDrive/PROJETOS/DASHBOARD_ADOLESCENTES_REPERTORIO/ADOLESCENTES - MUSICAS.csv')
```

```python
df_exec['ID_MUSICA'] = df_exec['ID_MUSICA'].str.strip()
```

```python
df_musicas = df_musicas.rename(columns={'ID_MÚSICA': 'ID_MUSICA'})
```

```python
df_final = pd.merge(df_exec, df_musicas, on='ID_MUSICA', how='left') #Esse final left, define a tabela em que todas as propriedades originais serão mantidas, poderia trocar por um right ou pela ordem em questão
```

```python
colunas_desejadas = [
    'ID_EXECUCAO',          # Para contar quantas vezes tocou no total
    'ID_MUSICA',            # Chave única (útil para debug)
    'MÚSICA',               # O nome oficial (Vem da tabela Músicas, mais confiável)
    'AUTOR_MUSICA',         # Para criar ranking de autores
    'CLASSIFICACAO_MUSICA', # Para o gráfico de Pizza (Agitada vs Lenta)
    'TOM',                  # Essencial para os músicos
    'STATUS'                # Para saber se a música está ativa
]

# Aplicando o filtro
df_limpo = df_final[colunas_desejadas]
```

```python
df_limpo.to_csv('DADOS_DASHBOARD_FINAL.csv', index=False)
```
